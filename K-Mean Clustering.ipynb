# **Read Dataset**


```python
#import library untuk membaca file csv dan visualisasi
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#membuat dataframe mahasiswa
dfmhs = pd.read_csv('nilaimahasiswa.csv')
```

```python
#mengecek karakter data
dfmhs.describe()
```

```python
#cek info karakter data
dfmhs.info()
```

```python
#visualisasi data
plt.scatter(dfmhs['uts'],dfmhs['uas'])
plt.xlabel("uts")
plt.ylabel("uas")
plt.title("Grafik Nilai Mahasiswa")
```

# **Membuat Data Training**

```python
#mengambil data training uts dan uas
x_train = dfmhs[['uts','uas']].values
x_train
```

# **Feature Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_train
```

# **Clustering**

```python
from sklearn.cluster import KMeans
kmean = KMeans(n_clusters=2)
kmean
```

```python
y_cluster = kmean.fit_predict(x_train)
y_cluster
```

```python
#menambah variable/aitribut baru untuk hasil cluster
dfmhs['cluster'] = y_cluster
dfmhs
```

```python
#mengetahui centroid
kmean.cluster_centers_
```

# **Visualisasi Data**

```python
#mengecek posisi data yg di cluster
plt.figure(figsize=(10,5))
plt.scatter(x_train[:,0], x_train[:,1], c=kmean.labels_)
plt.scatter(kmean.cluster_centers_[0,0], kmean.cluster_centers_[0,1], cmap='rainbow', marker='*', s=150, label='Centroid 0')
plt.scatter(kmean.cluster_centers_[1,0], kmean.cluster_centers_[1,1], cmap='rainbow', marker='*', s=150, label='Centroid 1')
#plt.scatter(kmean.cluster_centers_[2,0], kmean.cluster_centers_[2,1], cmap='rainbow', marker='*', s=150, label='Centroid 2')
plt.xlabel('UTS')
plt.ylabel('UAS')
plt.legend()
```

# **Mencari Jumlah Cluster Terbaik**

```python
inertias = []
k_range = range(1,10)
for k in k_range:
  km = KMeans(n_clusters=k).fit(x_train)
  inertias.append(km.inertia_)
inertias
```

```python
plt.xlabel("k")
plt.ylabel("Sum of error")
plt.plot(k_range,inertias)
plt.grid()
```

# **Membuat Dataset Baru dengan label cluster**

```python
dfmhs['cluster'] = dfmhs['cluster'].replace([0,1],['kurang','baik'])
dfmhs
```
